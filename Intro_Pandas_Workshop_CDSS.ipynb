### Joins/ Merges

Whats the difference? 
1. Merge does an inner join by default, while join does a left join.
2. Join by default joins on the index of the two dataframes.

```python
#Make sure you have an index/ID Column to ensure a correct result
df_titanic.index.rename('ID', inplace=True)
```

```python
passengers = df_titanic[['name','sex', 'age']]
```

```python
found_bodies = df_titanic[['body']]
```

```python
passengers.join(found_bodies)
```

```python
pd.merge(passengers,found_bodies, how='left', left_index=True, right_index=True)
```

```python
pd.merge(passengers, found_bodies, how='right',left_index=True, right_index=True)
```

### Set operations


Lets find the Union of two parts of the dataframe. Say we split the original df into those younger than 20 and those older.

```python
df_young = df_titanic.loc[df_titanic.age <=20]
df_old = df_titanic.loc[df_titanic.age >=20]
```

```python
#Union
df_all = pd.concat([df_young, df_old])
df_all
```

```python
#alternative using series union
df_titanic.loc[df_young.index.union(df_old.index)]
```

Now we split the age by younger than 30 and older than 20 and then find the intersection

```python
df_young2 = df_titanic.loc[df_titanic.age <=30]
df_old2 = df_titanic.loc[df_titanic.age >=20]
```

```python
#intersection
df_inter = pd.merge(df_young2, df_old2, how='inner', left_index=True, right_index=True)
```

```python
df_inter
```

```python
#alternative using series intersection
df_titanic.loc[df_young2.index.intersection(df_old2.index)]
```
